### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calculamos el ingreso total para cada tienda

primera_tienda = tienda['Precio'].sum()
segunda_tienda = tienda2['Precio'].sum()
tercera_tienda = tienda3['Precio'].sum()
cuarta_tienda = tienda4['Precio'].sum()

# Imprimirmos los resultados de los ingresos totales de las tiendas

print(f'El ingreso total para la tienda 1 es de: {primera_tienda:.2f}') # Usamos formato de especificación ": .2f" para que salga con dos decimales
print(f'El ingreso total para la tienda 2 es de: {segunda_tienda:.2f}')
print(f'El ingreso total para la tienda 3 es de: {tercera_tienda:.2f}')
print(f'El ingreso total para la tienda 4 es de: {cuarta_tienda:.2f}')
```

# 2. Ventas por categoría

```python
# tienda uno, como dataframe para guía de como agrupar
tienda[['Producto', 'Categoría del Producto', 'Precio']].head()
```

```python
# Agrupamos la cantidad de productos vendidos por categoria: groupby([])
agrupar_tienda_uno = tienda.groupby(['Categoría del Producto'])
agrupar_tienda_dos = tienda2.groupby(['Categoría del Producto'])
agrupar_tienda_tres = tienda3.groupby(['Categoría del Producto'])
agrupar_tienda_cuatro = tienda4.groupby(['Categoría del Producto'])

# Contamos numero de ventas por categorias: size()
categorias_tienda_uno = agrupar_tienda_uno.size()
categorias_tienda_dos = agrupar_tienda_dos.size()
categorias_tienda_tres = agrupar_tienda_tres.size()
categorias_tienda_cuatro = agrupar_tienda_cuatro.size()

print(f'Las ventas por categorías más populares de Tienda 1 son: {categorias_tienda_uno.sort_values(ascending=False)}') # .sort_values(ascending=False) permite ordenar de mayor a menor
print(f'Las ventas por categorías más populares de Tienda 2 son: {categorias_tienda_dos.sort_values(ascending=False)}')
print(f'Las ventas por categorías más populares de Tienda 3 son: {categorias_tienda_tres.sort_values(ascending=False)}')
print(f'Las ventas por categorías más populares de Tienda 4 son: {categorias_tienda_cuatro.sort_values(ascending=False)}')
```

# 3. Calificación promedio de la tienda


```python
# Calculamos las calificaciones promedios de los clientes para cada tienda: mean()
promedio_tienda_uno = tienda['Calificación'].mean()
promedio_tienda_dos = tienda2['Calificación'].mean()
promedio_tienda_tres = tienda3['Calificación'].mean()
promedio_tienda_cuatro = tienda4['Calificación'].mean()

print(f'La calificación de la tienda 1 es : {promedio_tienda_uno:.2f}')
print(f'La calificación de la tienda 2 es : {promedio_tienda_dos:.2f}')
print(f'La calificación de la tienda 3 es : {promedio_tienda_tres:.2f}')
print(f'La calificación de la tienda 4 es : {promedio_tienda_cuatro:.2f}')
```

# 4. Productos más y menos vendidos

```python
# Agrupamos por productos vendidos: .groupby() y contamos para saber mas y menos vendidos: .size()
productos_tienda_uno = tienda.groupby(['Producto']).size()
productos_tienda_dos = tienda2.groupby(['Producto']).size()
productos_tienda_tres = tienda3.groupby(['Producto']).size()
productos_tienda_cuatro = tienda4.groupby(['Producto']).size()

# Ordenamos por mas y menos vendidos: .sort_values()
mas_vendidos_tienda_uno = productos_tienda_uno.sort_values(ascending=False).head(10) # Mostrará los 15 más vendidos
menos_vendidos_tienda_uno = productos_tienda_uno.sort_values(ascending=True).head(10) # Mostrará los 10 menos vendidos

mas_vendidos_tienda_dos = productos_tienda_dos.sort_values(ascending=False).head(10)
menos_vendidos_tienda_dos = productos_tienda_dos.sort_values(ascending=True).head(10)

mas_vendidos_tienda_tres = productos_tienda_tres.sort_values(ascending=False).head(10)
menos_vendidos_tienda_tres = productos_tienda_tres.sort_values(ascending=True).head(10)

mas_vendidos_tienda_cuatro = productos_tienda_cuatro.sort_values(ascending=False).head(10)
menos_vendidos_tienda_cuatro = productos_tienda_cuatro.sort_values(ascending=True).head(10)

# mostramos los resultados
print(f'Los más vendidos Tienda Uno: {mas_vendidos_tienda_uno}')
print(f'Los más vendidos Tienda Uno: {menos_vendidos_tienda_uno}')
print(f'Los Productos más vendidos en la Tienda Dos fueron: {mas_vendidos_tienda_dos}')
print(f'Los Productos menos vendidos en la Tienda Dos fueron: {menos_vendidos_tienda_dos}')
print(f'Los Productos más vendidos en la Tienda Tres fueron: {mas_vendidos_tienda_tres}')
print(f'Los Productos menos vendidos en la Tienda Tres fueron: {menos_vendidos_tienda_tres}')
print(f'Los Productos más vendidos en la Tienda Cuatro fueron: {mas_vendidos_tienda_cuatro}')
print(f'Los Productos menos vendidos en la Tienda Cuatro fueron: {menos_vendidos_tienda_cuatro}')

# Resetamos el indice para obtener en columnas el producto y la cantidad de ventas en un dataframe, útil para luego graficar: .reset_index(name)
mas_vendidos_tienda_uno = mas_vendidos_tienda_uno.reset_index(name = 'Cantidad vendida')
menos_vendidos_tienda_uno = menos_vendidos_tienda_uno.reset_index(name= 'Cantidad vendida')
mas_vendidos_tienda_dos = mas_vendidos_tienda_dos.reset_index(name = 'Cantidad vendida')
menos_vendidos_tienda_dos = menos_vendidos_tienda_dos.reset_index(name= 'Cantidad vendida')
mas_vendidos_tienda_tres = mas_vendidos_tienda_tres.reset_index(name = 'Cantidad vendida')
menos_vendidos_tienda_tres = menos_vendidos_tienda_tres.reset_index(name= 'Cantidad vendida')
mas_vendidos_tienda_cuatro = mas_vendidos_tienda_cuatro.reset_index(name = 'Cantidad vendida')
menos_vendidos_tienda_cuatro = menos_vendidos_tienda_cuatro.reset_index(name= 'Cantidad vendida')
```

# 5. Envío promedio por tienda

```python
# Un vistaso al dataframe para identificar la columna respectiva
tienda.head()
```

```python
# Calculamos el promedio de envio por tienda
costo_envio_tienda_uno = tienda['Costo de envío'].mean()
costo_envio_tienda_dos = tienda2['Costo de envío'].mean()
costo_envio_tienda_tres = tienda3['Costo de envío'].mean()
costo_envio_tienda_cuatro = tienda4['Costo de envío'].mean()

# Mostramos resultados del gasto de envio por tienda

print(f'El gasto promedio en envío de la tienda uno es: {costo_envio_tienda_uno:.2f}')
print(f'El gasto promedio en envío de la tienda dos es: {costo_envio_tienda_dos:.2f}')
print(f'El gasto promedio en envío de la tienda tres es: {costo_envio_tienda_tres:.2f}')
print(f'El gasto promedio en envío de la tienda cuatro es: {costo_envio_tienda_cuatro:.2f}')
```

#### Comparación entre tiendas: Ingresos, Valoración de los clientes, costo de envío

```python
# Importamos la libreria para los graficos y el submodulo de ella

import matplotlib.pyplot as plt

# Agrupamos los nombres de las cuatro tiendas en una variable y los valores que vamos a comparar
# Eje X
tiendas_juntas = ['Tienda1', 'Tienda2', 'Tienda3', 'Tienda4']

# Ejes Y
ingresos_totales = [primera_tienda, segunda_tienda, tercera_tienda,cuarta_tienda]
calificaciones_tiendas = [promedio_tienda_uno, promedio_tienda_dos, promedio_tienda_tres, promedio_tienda_cuatro]
costos_de_envios = [costo_envio_tienda_uno, costo_envio_tienda_dos, costo_envio_tienda_tres, costo_envio_tienda_cuatro]

# Creamos el grafico de barras para Ingresos
plt.figure(figsize=(8, 5)) # Se crea una nueva figura de 8 pulgadas de ancho, y 5 de alto
plt.bar(tiendas_juntas, ingresos_totales)
plt.title('Ingresos Totales por tienda') # Colocar título
plt.xlabel('Tiendas')
plt.ylabel('Ingreso Total en Pesos ($)')
plt.show()

# Creamos el grafico de barras para Valoracion de clientes
plt.figure(figsize=(8, 5)) # Se crea una nueva figura y 8 pulgadas de ancho, y 5 de alto
plt.bar(tiendas_juntas, calificaciones_tiendas)
plt.title('Valoración de Clientes')
plt.xlabel('Tiendas')
plt.ylabel('Puntuación de Clientes')
plt.show()

# Creamos el grafico de barras para Costo de envío
plt.figure(figsize=(8, 5)) # Se crea una nueva figura y 8 pulgadas de ancho, y 5 de alto
plt.bar(tiendas_juntas, costos_de_envios)
plt.title('Promedio de Costos de envío')
plt.xlabel('Tiendas')
plt.ylabel('Costos en pesos ($)')
plt.show()
```
